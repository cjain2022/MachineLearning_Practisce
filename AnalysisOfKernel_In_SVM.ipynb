```python
from sklearn.datasets import make_circles 
import numpy as np 
from matplotlib import pyplot as plt
```

```python
X,Y=make_circles(n_samples=1000, noise =0.10)
```

```python
plt.figure(num=10,figsize=(10,10))
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
from sklearn import svm 
```

```python
# Linear Kernel 
svc=svm.SVC(kernel='linear')
svc.fit(X,Y)
svc.score(X,Y)
```

```python
# RBF Kernel 
svc=svm.SVC(kernel='rbf')
svc.fit(X,Y)
svc.score(X,Y)
```

```python
#Polynomial Kernel 
svc=svm.SVC(kernel='poly')
svc.fit(X,Y)
svc.score(X,Y)
```

```python
def customkernel(x1,x2):
    return np.square(np.dot(x1,x2.T))
svc=svm.SVC(kernel=customkernel)
svc.fit(X,Y)
svc.score(X,Y)
```
