```
#Using IMDB Dataset
from keras.datasets import imdb
((XT,YT),(Xt,Yt))=imdb.load_data(num_words=10000)
```

```
print(XT.shape)
print(YT.shape)
print(Xt.shape)
print(Yt.shape)
```

```
print(type(XT[0]),len(XT[0]))
print(type(Xt[0]),len(Xt[0]))
# It means that the datset is a 1 dimensional numpy array storing list 
print(type(YT[0]),Yt[0].shape)
print(type(Yt[0]),Yt[0].shape)
```

```
#Obtaining The mapping of numbers to words 
word_idx=imdb.get_word_index()
```

```
from keras.preprocessing import sequence
```

```
X_train=sequence.pad_sequences(XT,maxlen=500)
X_test=sequence.pad_sequences(Xt, maxlen=500)
```

```
print(type(X_train))
print(type(X_test))
print(XT.shape, len(XT[0]))
print(Xt.shape, len(Xt[0]))
print(X_train.shape)
print(X_test.shape)
```

```
from keras.models import Sequential 
```

```
model =Sequential()
```

```
from keras.layers import Embedding
model.add(Embedding(10000,64))
```

```
from keras.layers import SimpleRNN 
model.add(SimpleRNN(32))
```

```
from keras.layers import Dense
model.add(Dense(1,activation='sigmoid'))
```

```
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['acc'])
```

```
model.summary()
```

```
#model.fit(X_train,YT,batch_size=128,validation_split=0.2, epochs=10)
```

```
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```
checkpoint=ModelCheckpoint('IMDB_model.h5',monitor='val_loss',save_best_only=True, mode='auto',period=1)
StopCheckPoint=EarlyStopping(monitor='val_loss',patience=1,verbose=0, mode='auto')
```

```
hist=model.fit(X_train,YT,validation_split=0.2,epochs=20, batch_size=128, callbacks=[checkpoint,StopCheckPoint])
```

```
Y_pred=model.predict(X_test,batch_size=128)
```
